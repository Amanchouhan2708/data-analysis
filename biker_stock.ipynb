#Most Profitable Product

```python
brands.to_sql("brands",connection,if_exists='replace',index=False)
categories.to_sql("categories",connection,if_exists='replace',index=False)
customers.to_sql("customers",connection,if_exists='replace',index=False)
orders.to_sql("orders",connection,if_exists='replace',index=False)
order_items.to_sql("order_items",connection,if_exists='replace',index=False)
products.to_sql("products",connection,if_exists='replace',index=False)
staffs.to_sql("staffs",connection,if_exists='replace',index=False)
stocks.to_sql("stocks",connection,if_exists='replace',index=False)
stores.to_sql("stores",connection,if_exists='replace',index=False)
```

```python
query = """
    SELECT 
        oi.product_id,
        product_name,
        p.brand_id,
        SUM(quantity) as unit_sold,
        oi.list_price,
        discount,
        oi.list_price * (1 - discount) AS discounted_price,
        SUM(SUM((oi.list_price * (1 - discount))*quantity)) OVER (PARTITION BY oi.product_id)  AS total_sales_per_product
    FROM 
        order_items as oi
        JOIN products as p
            ON oi.product_id = p.product_id
    GROUP BY 1
    ORDER BY 8 DESC, 5 DESC
    LIMIT 10
        """
df = pd.read_sql_query(query, connection)
df
```

```python
order = df.sort_values('total_sales_per_product', ascending=False)['product_name']

fig, ax = plt.subplots(figsize=(22, 10), sharey=True)
sns.barplot(data=df, x='total_sales_per_product', y='product_name', orient='h' ,ax=ax,order=order, width = 0.6)

for p in ax.patches:
    width = p.get_width()
    ax.text(width + 0.1, p.get_y() + p.get_height() / 2, f'${width:,.0f}', 
            ha='left', va='center',fontsize=13)
    
ax.tick_params(axis='y', labelsize=14)    
ax.tick_params(axis='x', labelsize=14)    
ax.set_xlabel('Total Sales', fontsize=15)
ax.set_ylabel('Product Name', fontsize=15)
ax.set_title('Top 10 Most Profitable Product',fontsize=20)
plt.show()
```

```python

query = """

SELECT
    b.brand_name,
    COALESCE(SUM(oi.quantity), 0) AS total_quantity,
    COALESCE(ROUND(SUM(oi.list_price * (1 - oi.discount)), 2), 0) AS total_price
FROM 
    brands b
LEFT JOIN 
    products p ON b.brand_id = p.brand_id
LEFT JOIN 
    order_items oi ON p.product_id = oi.product_id
GROUP BY
    b.brand_name
ORDER BY
    total_price DESC;

"""
brands_df = pd.read_sql_query(query, connection)
brands_df
```

```python
plt.figure(figsize=(8,4))
ax = sns.barplot(data=brands_df, x='brand_name',y='total_price',color='darkseagreen')
ax.bar_label(ax.containers[0], labels=brands_df['total_quantity'])
plt.grid('x')
plt.xlabel("Brand Name")
plt.ylabel("Total Revenue (in millions)")
plt.title("Total Revenue by Brand with Units Solds")
plt.xticks(rotation = 45, horizontalalignment='right', fontsize=12)
plt.show()
```

```python
query = """

SELECT
    c.category_name,
    SUM(oi.quantity) AS total_quantity,
    ROUND(SUM(oi.list_price * (1 - oi.discount)), 2) AS total_price
FROM
    categories c
JOIN 
    products p ON c.category_id = p.category_id
JOIN 
    order_items oi ON p.product_id = oi.product_id
GROUP BY
    c.category_name
ORDER BY
    total_price DESC;


"""
categories_df = pd.read_sql_query(query, connection)
categories_df
```

```python
plt.figure(figsize=(8, 4))
ax = sns.barplot(data=categories_df, x='category_name', y='total_price', hue='category_name', dodge=False)
plt.grid('x')
plt.xlabel("Category Name")
plt.ylabel("Total Revenue (in millions)")
plt.title("Total Revenue by Category")
plt.xticks(rotation=90, fontsize=10)
plt.show()
```

```python

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.pie(categories_df['total_quantity'], labels=categories_df['category_name'], autopct='%1.1f%%', startangle=140)
plt.title('Total Quantity by Category')

# Create pie chart for total price
plt.subplot(1, 2, 2)
plt.pie(categories_df['total_price'], labels=categories_df['category_name'], autopct='%1.1f%%', startangle=140)
plt.title('Total Price by Category')

plt.tight_layout()
plt.show()
```
